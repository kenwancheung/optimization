## Bisection Method

The bisection method relies on the intermediate value theorem which states that if a function $f(x)$ is continuous on an interval $[a,b]$ and the sign of $f(a)$ differs from the sign of $f(b)$, then there exists a point $c$ in $[a,b]$ such that $f(c)=0$.

The figure below depicts several examples of continuous functions in a closed interval whose sign shift at an unknown point internal to the closed interval. The figure on the top left corner shows a function typically used in financial analysis. It shows the present value of a stream of coupon and principal payment as a function of the discount rate minus the borrowed amount. The root of this equation is the implied internal rate of return to the lender which equates the present valueof the coupon and principal payments with the borrowed amount. In this case, the internal rate of return is $2\%$. 

The bottom left figure presents the function $x^{2}-7$. The function has a negative root and a positive root. Depending on the interval range, the solution is $-2.64575$ or $2.64575$ - these are known as the positive and negative square root of $7$.

The top left figure shows a function that is not continuous and does not have a root. The bottom left figure shows a function that does not switch signs within the given interval range.


```julia
using PyPlot

x1 = collect(0.0:0.001:0.059)
f1 = 1./(1+x1).*20+(1./(1+x1).^2)*20+(1./(1+x1).^3)*20+(1./(1+x1).^4)*1020 - 1000

x2 = collect(0.0:0.05:2.0)
f2a = ifelse(x2.<1,exp(-x2)-exp(-1),NaN)
f2b = ifelse(x2.>1,exp(-x2)-exp(-0.8),NaN)

x3 = collect(-4.0:0.1:4.0)
f3 = x3.^2-7

x4 = collect(-1:0.01:1)
f4 = x4.^2+0.5

fig, ax = subplots(2,2)
ax[1,1][:plot](x1,f1,color="red",linestyle="-",linewidth=2,label=L"\frac{1}{1+x}20+\frac{1}{(1+x)^{2}}20+\frac{1}{(1+x)^{3}}20+\frac{1}{(1+x)^{4}}1020",alpha=1)
ax[1,1][:legend](loc="top center",fontsize=8)
ax[1,1][:plot](x1,zeros(length(x1)),color="black",linestyle="--",linewidth=2,label=L"f(x)=(x-1)^{2}",alpha=1)

ax[1,2][:plot](x2,f2a,color="red",linestyle="-",linewidth=2,label=L" ",alpha=1)
ax[1,2][:plot](x2,f2b,color="red",linestyle="-",linewidth=2,label=L" ",alpha=1)
ax[1,2][:plot](x2,zeros(length(x2)),color="black",linestyle="--",linewidth=2,label=L"f(x)=(x-1)^{2}",alpha=1)

ax[2,1][:plot](x3,f3,color="blue",linestyle="-",linewidth=2,label=L"x^2-7",alpha=1)
ax[2,1][:legend](loc="top center",fontsize=8)
ax[2,1][:plot](x3,zeros(length(x3)),color="black",linestyle="--",linewidth=2,label=L" ",alpha=1)

ax[2,2][:plot](x4,f4,color="blue",linestyle="-",linewidth=2,label=L"x^{2}+0.5",alpha=1)
ax[2,2][:legend](loc="top center",fontsize=8)
ax[2,2][:plot](x4,zeros(length(x4)),color="black",linestyle="--",linewidth=2,label="",alpha=1)

fig
```

This root finding method successively shrinks the starting interval $[a,b]$ to a sufficiently small range around the solution to $f(x)=0$. It starts by evaluating the function, $f(x)$ at the midpoint, $c = \frac{a+b}{2}$ and its sign (i.e., positive, negative or zero). Then the method replaces $a$ with the midpoint $c$ if the sign of $f(a)$ equals the sign of $f(c)$. Else, it replaces $b$ with the midpoint $c$ if the sign of $f(b)$ equals the sign of $f(c)$. It continues this process iteractively until 
the solution is within a interval with a pre-specified tolerance range.

```julia
using DataFrames

function ff(x)
    x^2-2
end

function bisec(f::Function,a::Float64,b::Float64,tol::Float64)
    if a>=b
        error("a must be less than b") 
    end
    if f(a)*f(b) > 0
        error("signs of f(a) and f(b) are not opposites") 
    end
    i = 1
    err=b-a
    c=(a+b)/2
    resual = DataFrame(a=a,b=b,c=c,err=err)
    while err>tol
        if f(c)==0.0
            break
        end
        if f(a)*f(c)<0
            a=a
            b=c
        else
            a=c
            b=b
        end
        c=(a+b)/2
        err=b-a
        i = i + 1
        push!(resual, @data([a,b,c,err]))
    end
    return c, resual
end
c, resual = bisec(ff,1.0,2.0,1e-6);
println("Bisection result")
println(c)
println("Julia's sqrt function")
println(sqrt(2))
println("Iterations")
println(resual)
```

Notice that the bisection method implies that the interval range always decreases by a factor of $2$. This implies that, for an initial interval range $\epsilon_0$, the number of iteration needed to arrive at an interval with range $\epsilon < \epsilon_0$ is $log_2 \frac{\epsilon_0}{\epsilon}$. The example above started with $a=1$ and $b=2$ implying an initial range of $1$, the tolerance level $\epsilon$ was set to $10^{-6}$. As a result, the number of iterations needed to arrive at the desired precision is $log_2 \frac{\epsilon_0}{\epsilon}=log_2 \frac{1}{10^{-6}}$ which is approximately $20$.

## Newton-Raphson method

This section discusses the Newton-Raphson method - an iterative algorithm helpful in finding the root of a non-linear equation or system of non-linear equations. It introduces the root find method and examples for single and multiple equations. This method plays an important role in many non-linear unconstraint optimization problems, including maximum likelihood estimation for linear and non-linear models.

### Root Find Method

In most situations, it is not possible to find analytical solutions for optimization problems. For example, maximum likelihood estimation for generalized linear models do not have closed-form solutions in most cases. As a result, numerical methods that iteratively search for the solution are necessary. This lecture begins with a few numerical examples.

The Newton-Raphson method is an algorithm used to find the root of a function or set of functions. The algorithm uses the first terms of a Taylor series in the vicinity of a potential root. The Taylor series of $f(x)$ about $x=x_{0}+\epsilon$ is:

$$
\begin{aligned}
	f(x) = f(x_{0}) + \frac{1}{1!}f'(x_{0}) \epsilon + \frac{1}{2!}f''(x_{0}) \epsilon^{2} + ... 
\end{aligned}
$$

Or approximately:

$$
\begin{aligned}
	f(x) \sim f(x_{0}) + f'(x_{0}) \epsilon
\end{aligned}
$$

This approximation represents the tangent line evaluated at $(x_{0},f(x_{0}))$. For a very small $\epsilon$, $-\frac{f(x_{0})}{f'(x_{0})}$ represents the change in $x_{0}$ or delta need for $f(x_{0}+\epsilon)=0$. To see that simply set $f(x)=0$ in the equation above and solve for $\epsilon$. The Newton-Raphson method is a root-finding method in mathematics used to solve for $f(x)=0$. First, one must make an initial guess for the solution, $x_{0}$. Then apply the algorithm below iteratively:

$$
\begin{aligned}
	x_{n+1} = x_{n} - \frac{f(x_{n})}{f'(x_{n})}
\end{aligned}
$$

Since we used only the first two terms of the Taylor expansion, the ``step'' ($-\frac{f(x_{0})}{f'(x_{0})}$) is only an approximation. Under regularity conditions, iterations of the algorithm gets us closer and closer to the solution.

The Newthon-Raphson method is not restricted to a single variable or single function. In fact, it can be applied to find the roots of a system of functions or equations with multiple variables. We will discuss an example later in this lecture. Let's first go over a simple example. Suppose we want to solve the following equation:

$$
\begin{aligned}
	f(x) = x^{2} - 2 = 0
\end{aligned}
$$

in the interval $[-2,2]$. 

The figure below plots the function $f(x)=x^{2}-2$ and shows where $f(x)=0$. The solution is $+/-\sqrt{2}=+/-1.41421...$. This is a non-linear equation and multiple techniques are available to solve it. We will describe a well-known technique to find the root of a non-linear equation called the Newton-Raphson method which can also be used to find the root of more complicated functions. It does not always work, but we will discuss its limitations later on.




```julia
using PyPlot

x = collect(-2:0.1:2)
ff = x.^2 - 2

fig, ax = subplots()
ax[:plot](x,ff,color="blue",linewidth=2,label=L"f(x) = x^{2} - 2",alpha=1)
ax[:legend](loc="top center")
ax[:plot](x,zeros(length(x)),color="black",linewidth=2,label=L"",alpha=0.3)
```

The program below applies the Newton-Raphson method to find the positive root of $f(x)=x^{2}-2$. The figure below depicts the iterative steps of the Newton-Raphson method for finding the root of $f(x)=x^{2}-2$. The initial guess is set to $`r x0`$. 


```julia
using DataFrames

tol = 10.0^(-5)
err = 10.0^5
x0 = 0.1

function f(x)
    x^2 - 2
end
function fp(x)
    2*x
end

k = 0
resual = DataFrame(k=NaN,x=NaN,f=NaN,fp=NaN,err=NaN)

while err>tol
    x = x0 - f(x0)/fp(x0)
    err = abs(x-x0)
    if k==0
        resual = DataFrame(k=k,x=x0,f=f(x0),fp=fp(x0),err=err)
    else
        push!(resual, @data([k,x0,f(x0),fp(x0),err]))
    end
    x0 = x
    k = k + 1
end
resual
```

```julia
fig, ax = subplots(2,2)

ax[1,1][:plot](resual[:k],resual[:x],color="red",linewidth=2,label="x",alpha=1)
ax[1,1][:legend](loc="upper right")
ax[1,2][:plot](resual[:k],resual[:f],color="red",linewidth=2,label="f",alpha=1)
ax[1,2][:legend](loc="upper right")
ax[2,1][:plot](resual[:k],resual[:fp],color="blue",linewidth=2,label=L"f'",alpha=1)
ax[2,1][:legend](loc="upper right")
ax[2,2][:plot](resual[:k],resual[:err],color="blue",linewidth=2,label=L"\epsilon",alpha=1)
ax[2,2][:legend](loc="upper right")
```

Notice that the process converges after $8$ iterations. The positive solution found is $r x$ which we know solves $f(x)=x^{2}-2$ approximately. The function $f(x)=x^{2}-2$ also has a negative solution. The algorithm can find the negative solution using a different starting point, say $x_{0}=-2$.


```julia
using DataFrames

tol = 10.0^(-5)
err = 10.0^5
x0 = -0.1

function f(x)
    x^2 - 2
end
function fp(x)
    2*x
end

k = 0
resual = DataFrame(k=NaN,x=NaN,f=NaN,fp=NaN,err=NaN)

while err>tol
    x = x0 - f(x0)/fp(x0)
    err = abs(x-x0)
    if k==0
        resual = DataFrame(k=k,x=x0,f=f(x0),fp=fp(x0),err=err)
    else
        push!(resual, @data([k,x0,f(x0),fp(x0),err]))
    end
    x0 = x
    k = k + 1
end
resual
```

```julia
fig, ax = subplots(2,2)

ax[1,1][:plot](resual[:k],resual[:x],color="red",linewidth=2,label="x",alpha=1)
ax[1,1][:legend](loc="lower right")
ax[1,2][:plot](resual[:k],resual[:f],color="red",linewidth=2,label="f",alpha=1)
ax[1,2][:legend](loc="upper right")
ax[2,1][:plot](resual[:k],resual[:fp],color="blue",linewidth=2,label=L"f'",alpha=1)
ax[2,1][:legend](loc="lower right")
ax[2,2][:plot](resual[:k],resual[:err],color="blue",linewidth=2,label=L"\epsilon",alpha=1)
ax[2,2][:legend](loc="upper right")
```

It is important to understand how the Newton-Raphson method works and be aware of its limitations. The Newton-Raphson method will not always find the same solution across different starting values. As the example above showed, the solution found for an initial value greater than $0$ (i.e., argmin of $f(x)$) was positive while the solution found for an initial value less than $0$ was negative. It is critical to choose well your initial values and to understand the underlying function.

## Multiple Variables

For GLM models, we will be looking for solutions to multiple optimization equations. A more realistic example would have two variables and two equations. Consider the following vector with two functions:

$$
\begin{aligned}
		f(x) = \left[\begin{array}{c}
				 	 f_{1}(x_{1},x_{2}) \\\\
					 f_{2}(x_{1},x_{2})
				 \end{array}\right] = \left[\begin{array}{c}
				 	 log(x_{1}^{2} + 2 x_{2}^{2} + 1) - 0.5 \\\\
					 x_{2} - x_{1}^{2} + 0.2
				 \end{array}\right]
\end{aligned}
$$

Our goal is to find the solution $x=(x_{1},x_{2})$ such that $f(x)=0$. Let's first plot $f_{1}(x)$. The figure below shows $f_{1}(x)$ for a grid $x$. By examining the plot we can guess that there is a range of $x$ for which $f_{1}(x)=0$. In fact, $f_{1}(x)=0$ if $x_{1}^{2}+2x_{2}^{2}=e^{0.5}-1$ or $x_{2}=\left[\frac{e^{0.5}-1-x_{1}^{2}}{2}\right]^{0.5}$. The figure below also shows $f_{2}(x)$ for a grid $x$. By examining the plot we can guess that there is a range of $x$ for which $f_{2}(x)=0$. In fact, $f_{2}(x)=0$ if $x_{1}^{2}-x_{2}=0.2$ or $x_{2}=x_1^{2}-0.2$.




```julia
using Plots
pyplot()

n = 50
x1 = collect(linspace(-1, 1, n))
x2 = collect(linspace(-1,1,n))

f1 = zeros(n,n)
f2 = zeros(n,n)

for i in 1:n
    for j in 1:n
        f1[i,j] = log(x1[i]^2+2*x2[j]^2+1)-0.5
    end
end

for i in 1:n
    for j in 1:n
        f2[i,j] = x2[j]-x1[i]^2+0.2
    end
end

default(size=(400,500))
Plots.plot(Plots.surface(x1,x2,f1,alpha=1,cbar=true,fc=:heat,c=:black),
     Plots.contour(x1,x2,f1),
     Plots.surface(x1,x2,f2,alpha=1,cbar=true,fc=:heat,c=:black),
     Plots.contour(x1,x2,f2),
     layout=@layout([a b; c d]),size=(1000,1000))
```

The plot below shows combinations for $x_{1}$ and $x_{2}$ such that $f_{1}(x)=0$ (red curve) and $f_{2}(x)=0$ (blue curve). Both functions will be simultaneously equal to zero at the intersection the two curves. So visually we can tell that there are two possible solution within the range of values for $x_{1}$ and $x_{2}$.

```julia
using PyPlot

n = 50
x1 = collect(linspace(-1, 1, n))

x2a = zeros(n,1)
for i=1:n
    if exp(0.5)-1-x1[i]^2>=0
        x2a[i] = ((exp(0.5)-1-x1[i]^2)/2)^0.5
    else
        x2a[i] = NaN
    end
end

x2b = x1.^2 - 0.2

fig, ax = subplots()
ax[:plot](x1,x2a,color="red",linewidth=2,label=L"x_{1} = \frac{e^{\frac{1}{2}}-1-x_{1}^{2}}{2}",alpha=1)
ax[:legend](loc="upper right")
ax[:plot](x1,x2b,color="blue",linewidth=2,label=L"x_{2} = x^{2} - 0.2",alpha=1)
ax[:legend](loc="upper right")

```

Based on the Newton-Raphson method we discussed above, we can intuitively guess what the multivariate algorithm would look like:

$$
\begin{aligned}
		x_{n+1} = x_{n} +  Df(x_{n})^{-1} f(x_{n})
\end{aligned}
$$

where 

$$
\begin{aligned}
		Df(x_{n}) = \left[\begin{array}{cc}
						\frac{\partial f_{1}(x_{n,1},x_{n,2})}{\partial x_{1}} & \frac{\partial f_{1}(x_{n,1},x_{n,2})}{\partial x_{2}}  \\\\
						\frac{\partial f_{2}(x_{n,1},x_{n,2})}{\partial x_{1}} & \frac{\partial f_{2}(x_{n,1},x_{n,2})}{\partial x_{2}}  
					\end{array}\right]
				  =
					\left[\begin{array}{cc}
						2 x_{n,1}/(x_{n,1}^{2}+2x_{n,2}^{2}+1) & 4 x_{n,2}/(x_{n,1}^{2}+2 x_{n,2}^{2}+1) \\\\
					   -2 x_{n,1} & 1
					\end{array}\right]
\end{aligned}
$$

After choosing a tolerance level and an initial set of values for $x$, we can run the algorithm below to find the solution to the system of equations $f(x)=0$.


```julia
using DataFrames

tol = 10.0^(-10)
err = 10.0^5
x0 = [1.0; 1.0]

x = copy(x0)

function f1(x)
    [log(x[1]^2+2*x[2]^2+1)-0.5; x[2]-x[1]^2+0.2]
end

function Df1(x)
    [2*x[1]/(x[1]^2+2*x[2]^2+1) 4*x[2]/(x[1]^2+2*x[2]^2+1);
        -2*x[1] 1]
end

k = 0
resual = DataFrame(k=NaN,x=NaN,f=NaN,invDf1=NaN,invDf2=NaN,err=NaN)

while err>tol
    x = x0 - inv(Df1(x0))*f1(x0)
    err = maximum(abs(x-x0))
    if k==0
#        resual = DataFrame(k=[k;k],x=x0,f=f1(x0),invDf1=inv(Df1(x0))[:,1],invDf2=inv(Df1(x0))[:,1],err=x-x0)
        resual = DataFrame(k=k,x=x0[1,1],f=f1(x0)[1,1],invDf1=inv(Df1(x0))[1,1],invDf2=inv(Df1(x0))[1,2],err=x[1,1]-x0[1,1])
        push!(resual, @data([k,x0[2,1],f1(x0)[2,1],inv(Df1(x0))[2,1],inv(Df1(x0))[2,2],x[2,1]-x0[2,1]]))
    else
#        push!(resual, @data([[k,k] x0 f(x0) inv(Df(x0)) x-x0]))
        push!(resual, @data([k,x0[1,1],f1(x0)[1,1],inv(Df1(x0))[1,1],inv(Df1(x0))[1,2],x[1,1]-x0[1,1]]))
        push!(resual, @data([k,x0[2,1],f1(x0)[2,1],inv(Df1(x0))[2,1],inv(Df1(x0))[2,2],x[2,1]-x0[2,1]]))
    end
    x0 = x
    k = k + 1
end
resual
```

## Golden Section Search

This section discusses the golden section search which is a derivative-free method to find the local minimum of a function in a given range. The function must be unimodal in the range of interest - i.e., only one minimal.

The method begins by braketing the minimum with a triplet $(a,b,c)$ where $a<b<c$ and $f(b)<min\{f(a),f(c)\}$. The figure below shows examples of properly and improperly brackets minima. The figure on the top left and right corners show properly bracketed minima. The blue, green, and red dots represents the coordinate $(a,f(a))=(0,0.5)$, $(b,f(b))=(0.5,0.1487)$, and $(c,f(c))=(1,0.2183)$, respectively. The figure on the bottom left and right corners show improperly bracketed minima, where $f(b)>f(c)$.


```julia
using PyPlot

x = collect(0.0:0.01:1)
f = exp(x)-2*x-0.5
a = 0
b = 1/2
c = 1

fa = exp(a)-2*a-0.5
fb = exp(b)-2*b-0.5
fc = exp(c)-2*c-0.5

fig, ax = subplots(2,2)
ax[1,1][:plot](x,f,color="gray",linestyle="-",linewidth=2,label=L"e^{x}-2x",alpha=1)
ax[1,1][:legend](loc="top center",fontsize=8)
ax[1,1][:plot](x,zeros(length(x)),color="black",linestyle="--",linewidth=2,label=L"",alpha=1)

ax[1,1][:scatter]([a],[fa],color="blue")
ax[1,1][:vlines](a,0,fa,color="blue")

ax[1,1][:scatter]([b],[fb],color="green")
ax[1,1][:vlines](b,0,fb,color="green")

ax[1,1][:scatter]([c],[fc],color="red")
ax[1,1][:vlines](c,0,fc,color="red")

a = 0.2
b = 0.8
c = 1

fa = exp(a)-2*a-0.5
fb = exp(b)-2*b-0.5
fc = exp(c)-2*c-0.5

ax[1,2][:plot](x,f,color="gray",linestyle="-",linewidth=2,label=L"e^{x}-2x",alpha=1)
ax[1,2][:legend](loc="top center",fontsize=8)
ax[1,2][:plot](x,zeros(length(x)),color="black",linestyle="--",linewidth=2,label=L"",alpha=1)

ax[1,2][:scatter]([a],[fa],color="blue")
ax[1,2][:vlines](a,0,fa,color="blue")

ax[1,2][:scatter]([b],[fb],color="green")
ax[1,2][:vlines](b,0,fb,color="green")

ax[1,2][:scatter]([c],[fc],color="red")
ax[1,2][:vlines](c,0,fc,color="red")

a = 0
b = 1/3
c = 1/2

fa = exp(a)-2*a-0.5
fb = exp(b)-2*b-0.5
fc = exp(c)-2*c-0.5

ax[2,1][:plot](x,f,color="gray",linestyle="-",linewidth=2,label=L"e^{x}-2x",alpha=1)
ax[2,1][:legend](loc="top center",fontsize=8)
ax[2,1][:plot](x,zeros(length(x)),color="black",linestyle="--",linewidth=2,label=L"",alpha=1)

ax[2,1][:scatter]([a],[fa],color="blue")
ax[2,1][:vlines](a,0,fa,color="blue")

ax[2,1][:scatter]([b],[fb],color="green")
ax[2,1][:vlines](b,0,fb,color="green")

ax[2,1][:scatter]([c],[fc],color="red")
ax[2,1][:vlines](c,0,fc,color="red")

a = 0
b = 1/3
c = 0.8

fa = exp(a)-2*a-0.5
fb = exp(b)-2*b-0.5
fc = exp(c)-2*c-0.5

ax[2,2][:plot](x,f,color="gray",linestyle="-",linewidth=2,label=L"e^{x}-2x",alpha=1)
ax[2,2][:legend](loc="top center",fontsize=8)
ax[2,2][:plot](x,zeros(length(x)),color="black",linestyle="--",linewidth=2,label=L"",alpha=1)

ax[2,2][:scatter]([a],[fa],color="blue")
ax[2,2][:vlines](a,0,fa,color="blue")

ax[2,2][:scatter]([b],[fb],color="green")
ax[2,2][:vlines](b,0,fb,color="green")

ax[2,2][:scatter]([c],[fc],color="red")
ax[2,2][:vlines](c,0,fc,color="red")
```

Given a proper bracket $(a,b,c)$, the next step is to find a new point $x$ either between $a$ and $b$ or between $b$ and $c$. Suppose we choose $x$ between $b$ and $c$, then the next step is to evaluate the function $f(x)$. If $f(x)>f(b)$, then the new bracket is $(a,b,x)$ - since the bracket $(a,b,x)$ is proper and narrower than $(a,b,c)$. Otherwise, the new bracket is $(b,x,c)$ - since $(b,x,c)$ is proper and narrower than $(a,b,c)$. 

Suppose now that we choose $x$ between $a$ and $b$ instead. Then the next step is to evaluate the function $f(x)$. If $f(x)>f(b)$, then the new bracket is $(x,b,c)$ - since the bracket $(x,b,c)$ is proper and narrower than $(a,b,c)$. Otherwise, the new bracket is $(a,x,b)$ - since $(a,x,b)$ is proper and narrower than $(a,b,c)$. This bracketing process continues until the range between the outer points are within a prespecified tolerance level.


The missing piece is how to choose $x$ given a bracket $(a,b,c)$. Without loss of generality, assume that $x$ falls between $b$ and $c$. Denote $w$ ratio of the distance between $a$ and $b$ to the distance between $a$ and $c$:

$$
\begin{align}
\frac{b-a}{c-a}&=w\\
\end{align}
$$

It is trivial to see that $1-w$ represents the ratio of the distance between $b$ and $c$ to the distance between $a$ and $c$:

$$
\begin{align}
\frac{c-b}{c-a}&=1-w\\
\end{align}
$$

Also, assume the next $x$ is a fraction $z$ beyond $b$:

$$
\begin{align}
\frac{x-b}{c-a}&=z\\
\end{align}
$$

Recall that if $f(b)<f(x)$ then the new bracket is $(a,b,x)$, otherwise the new bracket is $(b,x,c)$. In the former case, the ratio of the new bracket to the old bracket is $w+z$ while in the latter is $1-w$. Equating these two distances implies:

$$
\begin{align}
w+z&=1-w\\
  z&=1-2w\\
\end{align}
$$


For simplicity and without loss of generality, assume for the moment that $c-a=1$. Notice that:

$$
\begin{align}
c-x&=(c-b)-(x-b) \\
   &=1-w-z\\
   &=w\\
   &=b-a
\end{align}
$$

In words, the distance between $a$ and $b$ mirrors the distance between $c$ and $x$. 


Notice that choosing $x$ is equivalent to choosing $z$ since $x=b+z(c-a)$. To select $z$ appropriately we assume that we choose $w$ appropriately in the previous iteration. More specifically, we assume that the distance $x$ to $b$ relative to the distance $b$ to $c$ is the same as the distance $a$ to $b$ relative to the distance $a$ to $c$. That is:

$$
\begin{align}
\frac{x-b}{c-b}&=\frac{b-a}{c-a}\\
\frac{\frac{x-b}{c-a}}{\frac{c-b}{c-a}}&=\frac{b-a}{c-a}\\
\frac{z}{1-w}&=w\\
\end{align}
$$


Combining the last two equivalences yields (i.e., $z=1-2w$ or $\frac{z}{1-w}=w$):

$$
\begin{align}
w^2-3w+1&=0\\
\end{align}
$$
implying

$$
\begin{align}
w &=\frac{3-\sqrt{5}}{2}\\
  &=0.38197
\end{align}
$$


The figure below depicts the first $3$ steps of the golden section search algorithm applied to the function $f(x)=e^{x}-2x-\frac{1}{2}$ in the range $[0,1]$. The top left corner figure shows the set up of the initial bracketing and the value of $x$. We first set $a=0$, and $c=1$ and complete the bracketing of the minimum by setting $b=a+w(c-a)$, where $w=\frac{3-\sqrt{5}}{2} \approx 0.38197$ - i.e., $b \approx 0+0.38197*(1-0)=0.38197$. To determine the next bracket, we first set the value for $x$ and evaluate $f(b)$ and $f(x)$. First, we set $x=b+z(c-a)$, where $z=1-2w$ - i.e., $z \approx 1-2*0.38197=0.23606$ and $x \approx 0.38197+0.23606*(1-0)=0.61803$. Notice that $c-x=0.38197$. Since $f(b)>f(x)$, we update the bracket to $(b,x,c)$ - i.e., $(a,b,c)=(0.38197,0.61803,1)$. The upper right, lower left, and lower right figures represents, respectively, the 2nd, 3rd and 4th iterations of the algorithm.

```julia
using Optim, DataFrames

function h(x)
    exp(x)-2.0*x-0.5
end

# Given a function f and a bracket ax, bx, cx, search for the minimum
#function goldsec(f::Function,a::Float64,b::Float64,c::Float64,tol::Float64)

    C=(3.0-sqrt(5.0))/2.0;
    R=1-C;

    ax = 0.0
    cx = 1.0
    bx = ax+C*(cx-ax)
    
    x0=ax;
    x3=cx;

    if (abs(cx-bx) > abs(bx-ax))
        x1=bx;
        x2=bx+C*(cx-bx);
    else
        x2=bx;
        x1=bx-C*(bx-ax);
    end

    h1=h(x1);
    h2=h(x2);

#define SHFT2(a,b,c) (a)=(b);(b)=(c);
#define SHFT3(a,b,c,d) (a)=(b);(b)=(c);(c)=(d);
    tol=1.0e-6;
    k=0;
    resual = DataFrame(k=k,x0=x0,x1=x1,x2=x2,x3=x3,h1=h1,h2=h2);
    while (abs(x3-x0) > tol*(abs(x1)+abs(x2)))
        if (h2 < h1)
            x0=x1
            x1=x2
            x2=R*x1+C*x3
            h1=h2
            h2=h(x2)
        else
            x3=x2
            x2=x1
            x1=R*x2+C*x0
            h2=copy(h1)
            h1=copy(h(x1))
        end
        k=k+1
        push!(resual, @data([k,x0,x1,x2,x3,h1,h2]))
    end
#    if (f1 < f2)
 #       xmin=x1
  #      fmin=f1
   # else
    #    xmin=x2
     #   fmin=f2
    #end
#end
resual
```

## Steepest Descent Method

This section shows a simple example of the steepest descent method. Suppose we would like to solve the following minimization problem:

$$
\begin{align}
\min_x \quad x_1^2+x_2^2+2x_1+4 \\
\end{align}
$$

```julia
using Optim, DataFrames

## Steepest Descent Method

function f(x)
    return x[1]^2+x[2]^2+2*x[1]+4
end

function df(x)
    return [2*x[1]+2;2*x[2]]
end

x0 = [3.0;-2.0]

err = 1.0e8
tol = 1.0e-8

resual = DataFrame(k=NaN,x0=NaN,x=NaN,h0=NaN,err=NaN)
k=0
while err>tol
    # use golden search method, newton, or solve analytically
    opt = Optim.optimize(h -> f(x0+h*df(x0)),-3.0,3.0,GoldenSection())
    h0 = opt.minimizer
    # update the value of x
    x = x0+h0*df(x0)
    err = maximum(abs(x-x0))
    if k==0
        resual = DataFrame(k=[k;k],x0=x0,x=x,h0=[h0;h0],err=[err;err])
    else
        push!(resual, @data([k,x0[1],x[1],h0,err]))
        push!(resual, @data([k,x0[2],x[2],h0,err]))
    end
    x0 = x
    k = k + 1
end
resual
```

Next, we provide another example from Ch. 5 in CT.

$$
\begin{align}
\min_x \quad (x_1-2)^4+e^{x_1-2}+(x_1-2x_2)^2 \\
\end{align}
$$

```julia
using Optim, DataFrames

## Steepest Descent Method

function f(x)
    return (x[1]-2)^4+exp(x[1]-2)+(x[1]-2*x[2])^2
end

function df(x)
    return [4*(x[1]-2)^3+exp(x[1]-2)+2*(x[1]-2*x[2]);-4*(x[1]-2*x[2])]
end

x0 = [0.0;3.0]

err = 1.0e8
tol = 1.0e-3

resual = DataFrame(k=NaN,x0=NaN,x=NaN,h0=NaN,err=NaN)
k=0
while err>tol
    # use golden search method, newton, or solve analytically
    opt = Optim.optimize(h -> f(x0+h*df(x0)),-3.0,3.0,GoldenSection())
    h0 = opt.minimizer
    # update the value of x
    x = x0+h0*df(x0)
    err = maximum(abs(x-x0))
    if k==0
        resual = DataFrame(k=[k;k],x0=x0,x=x,h0=[h0;h0],err=[err;err])
    else
        push!(resual, @data([k,x0[1],x[1],h0,err]))
        push!(resual, @data([k,x0[2],x[2],h0,err]))
    end
    x0 = x
    k = k + 1
end
resual
```
